```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# PCA, MDS, K-Means
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import euclidean_distances, cosine_distances
from sklearn.manifold import MDS
from sklearn.cluster import KMeans
```

### Data Loading and Processing

The first step is to gather and clean the raw statistics for midfielders from FBref. We extract separate tables for shooting, passing, progression, and defensive actions. Many of these tables use MultiIndex headers, so we select the second level and remove any repeated header rows. Non-numeric or missing ‘Rk’ values are dropped to ensure clean indexing.

After cleaning, the individual tables are merged into a single consolidated dataframe keyed by player and team. This combined dataset contains all relevant features for each midfielder, including offensive, defensive, passing, and progression metrics.

```python
# Read the table and get second level of MultiIndex, drop header
shooting = pd.read_html("shooting.html", encoding="utf-8")[0]
shooting.columns = shooting.columns.get_level_values(1)
shooting = shooting[shooting['Rk'] != 'Rk'].reset_index(drop=True)
shooting['Rk'] = pd.to_numeric(shooting['Rk'], errors='coerce')
shooting = shooting.dropna(subset=['Rk'])
shooting.columns, len(shooting.columns)
```

```python
possession = pd.read_html("possession.html", encoding="utf-8")[0]
possession.columns = possession.columns.get_level_values(1)
possession = possession[possession['Rk'] != 'Rk'].reset_index(drop=True)
possession['Rk'] = pd.to_numeric(possession['Rk'], errors='coerce')
possession = possession.dropna(subset=['Rk'])
possession.columns, len(possession.columns)
```

```python
passing = pd.read_html("passing.html", encoding="utf-8")[0]
passing.columns = passing.columns.get_level_values(1)
passing = passing[passing['Rk'] != 'Rk'].reset_index(drop=True)
passing['Rk'] = pd.to_numeric(passing['Rk'], errors='coerce')
passing = passing.dropna(subset=['Rk'])
passing.columns, len(passing.columns)
```

```python
creation = pd.read_html("creation.html", encoding="utf-8")[0]
creation.columns = creation.columns.get_level_values(1)
creation = creation[creation['Rk'] != 'Rk'].reset_index(drop=True)
creation['Rk'] = pd.to_numeric(creation['Rk'], errors='coerce')
creation = creation.dropna(subset=['Rk'])
creation.columns, len(creation.columns)
```

```python
defence = pd.read_html("defence.html", encoding="utf-8")[0]
defence.columns = defence.columns.get_level_values(1)
defence = defence[defence['Rk'] != 'Rk'].reset_index(drop=True)
defence['Rk'] = pd.to_numeric(defence['Rk'], errors='coerce')
defence = defence.dropna(subset=['Rk'])
defence.columns, len(defence.columns)
```

Now that each individual table has been cleaned, we merge them into a single dataframe keyed by player name. This results in a larger dataset containing all relevant statistics for each midfielder. During the merge, duplicate entries are dropped to ensure that each player appears only once. Finally, we check the size of the consolidated dataframe to confirm it matches the expected number of players and features, ensuring that the merge was successful.

The goal of this preprocessing is to produce a clean, structured dataset that allows for comprehensive analysis of player performance, dimensionality reduction, clustering, and similarity comparisons.

```python
from functools import reduce

# 5 dataframes
dfs = [shooting, possession, passing, creation, defence]

seen_cols = set()
cleaned_dfs = []
for df in dfs:
    keep_cols = ['Player'] + [col for col in df.columns if col != 'Player' and col not in seen_cols]
    seen_cols.update(df.columns.difference(['Player']))
    cleaned_dfs.append(df[keep_cols])

# Merge all DFs on 'Player'
merged_df = reduce(lambda left, right: pd.merge(left, right, on='Player', how='outer'), cleaned_dfs)

# Drop duplicates
def drop_duplicate_columns(df):
    df_T = df.T.drop_duplicates()
    return df_T.T
merged_df = drop_duplicate_columns(merged_df)
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]
merged_df = merged_df.drop_duplicates(subset='Player', keep='first')
merged_df['90s'] = pd.to_numeric(merged_df['90s'], errors='coerce')
merged_df['Age'] = pd.to_numeric(merged_df['Age'], errors='coerce')


print(f"Shape: {merged_df.shape}")
merged_df.columns
```

```python
# Recover datatframe with midfielders
mids = merged_df[(merged_df['Pos'] == 'MF') | (merged_df['Pos'] == 'MF,DF')]
mids.head()
```

After obtaining the full dataset with all featues, we now select the features relevant for our analysis, including offensive, passing, progression, defensive, and error metrics. To ensure consistency and comparability between players, all numeric features are converted to numeric types and missing values are filled with zeros. We filter out players with fewer than 10 90-minute appearances to focus on those with sufficient playing time. Percentage-based statistics are left as-is, while all other metrics are normalised per 90 minutes to account for differences in playing time. This produces a final cleaned and standardised dataframe to be exported, which is used for streamlit integration.

```python
# Metadata + feature columns
metadata_cols = ['Player', 'Nation', 'Squad', 'Comp', 'Age', '90s']
features = [
    'Gls', 'Ast', 'xG', 'npxG', 'xA', 'SCA', 'Sh', 'SoT', 'SoT%',
    'Cmp', 'Cmp%', 'PrgP', 'KP', 'PPA', 'CrsPA',
    'Carries', 'PrgC', 'PrgR', 'CPA', 'PrgDist',
    'Tkl', 'TklW', 'Int', 'Blocks', 'Mis', 'Dis', 'Fld', 'Err'
]
mids_selected = mids[metadata_cols + features].copy()

# Convert all feature columns to numeric, fill NaNs with 0
mids_selected[features] = mids_selected[features].apply(pd.to_numeric, errors='coerce')
mids_selected[features] = mids_selected[features].fillna(0)

percent_features = ['SoT%', 'Cmp%']
per90_features = [f for f in features if f not in percent_features]
mids_selected = mids_selected[mids_selected['90s'] > 10] # Keep if more than 10 90s

# Normalise features per 90
mids_selected[per90_features] = mids_selected.apply(
    lambda row: row[per90_features] / row['90s'],
    axis=1
)
# Export final dataframe for streamlit app
mids_selected.to_csv("mids_selected.csv", index=False, encoding="utf-8") 
```

We have obtained useful features on a per 90 scale as required.

### Correlation and Clustered Correlation heatmaps
Next we look at the correlations between the different features in our dataset. The heatmap shows how certain metrics move together, for example goals, expected goals, shots and other goal threat metrics are closely related, while defensive actions like tackles and interceptions are more strongly connected to each other while being very independent of all other statistics.

We also create a clustered heatmap, which groups features with similar correlation patterns. This makes it easier to spot redundancy in the data and gives us a sense of which statistics provide overlapping information. These insights are useful to keep in mind for the next step, where we apply PCA for dimensionality reduction.

```python
corr = mids_selected[features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Correlation Heatmap of Midfielder Features", fontsize=14)
plt.show()

sns.clustermap(
    corr,
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    figsize=(6, 6)
)

plt.suptitle("Clustered Correlation Heatmap of Midfielder Features", fontsize=14, y=1.02)
plt.show()
```

### PCA Components and Plot
We apply Principal Component Analysis (PCA) to reduce the dimensionality of our dataset and better understand the structure of the features. The cumulative variance plot shows that the first two principal components explain only about 50% of the total variance, which means a lot of information is still spread across higher dimensions.

Plotting the players on the first two principal components does not reveal very clear patterns. Many points overlap with one another, and distinct clusters are hard to identify. Looking at the loadings, the first component is influenced most by creative and attacking metrics such as shot-creating actions, progressive receptions, expected assists and key passes, while the second component is driven more by passing and ball retention features like pass completion, carries and progressive passing distance.

Since the two PCA dimensions do not capture enough of the variance and do not produce a meaningful separation of players, we turn to Multidimensional Scaling (MDS). To visualise player similarity, we need a two-dimensional representation of the data that captures as much of the underlying structure as possible. MDS allows us to do this by projecting players into two dimensions while preserving the pairwise distances between them. This approach is better suited than PCA for our purposes, since it retains more of the information from higher dimensions. The resulting two-dimensional MDS plot provides a clear basis for visualising clusters of similar midfielders and for highlighting relationships between players in an interpretable way.

```python
# Extract feature matrix and standardise for PCA
X = mids_selected[features].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Fit PCA with all components, plot variance by no. of components
pca_full = PCA().fit(X_scaled)
explained_var = pca_full.explained_variance_ratio_
plt.figure(figsize=(9,6))
plt.plot(range(1, len(explained_var)+1), np.cumsum(explained_var), marker='o')
plt.xticks(range(1, len(explained_var)+1))
plt.xlabel("Principal Component")
plt.ylabel("Cumulative Variance Explained")
plt.title("Cumulative Explained Variance by PCA")
plt.axhline(0.8, color='r', linestyle='--', label="80% threshold")
plt.legend()
plt.show()

# Plot PCA with 2 components for visualisation
pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(9, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7)
plt.title("PCA of Midfielder Stats (Top 5 Leagues)", fontsize=14)
plt.xlabel("PC1 (%.1f%% variance)" % (pca.explained_variance_ratio_[0] * 100))
plt.ylabel("PC2 (%.1f%% variance)" % (pca.explained_variance_ratio_[1] * 100))
plt.show()

# Variance explained by 2 components
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained:", pca.explained_variance_ratio_.sum())

loadings = pd.DataFrame(
    pca.components_.T,
    index=features,
    columns=['PC1', 'PC2']
)

# View prominent features for first 2 principal components
top_PC1 = loadings['PC1'].abs().sort_values(ascending=False).head(10)
top_PC2 = loadings['PC2'].abs().sort_values(ascending=False).head(10)

print("Top contributing features to PC1:\n", top_PC1, "\n")
print("Top contributing features to PC2:\n", top_PC2)
```

### MDS
To analyse player similarity, we start by projecting the selected features into a lower-dimensional space using PCA, retaining around 12 components to capture most of the variance. Using the PCA projections instead of the raw features helps reduce noise and correlations between metrics, providing more stable and meaningful distances between players.

From the PCA projections, we calculate pairwise distances using both Euclidean and cosine metrics. Euclidean distance measures overall differences in feature values, while cosine distance captures similarity in the pattern of a player’s features regardless of magnitude. These distance matrices serve as the input for Multidimensional Scaling (MDS), which will allow us to visualise the players in two dimensions.

We also apply K-means clustering on the PCA projections, grouping players into four clusters based on similarity. These cluster assignments are later used to colour points in the MDS plots, highlighting groups of players with similar playing styles. Finally, the top 25 players by minutes are annotated so that the plots are easier to interpret and to give context to the most active midfielders.

```python
def compute_embeddings(df, n_clusters=4):
    X = df[features].values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    pca = PCA(n_components=0.9)
    X_pca = pca.fit_transform(X_scaled)

    dist_euc = euclidean_distances(X_pca)
    dist_cos = cosine_distances(X_pca)

    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(X_pca)

    mds_euc = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
    coords_euc = mds_euc.fit_transform(dist_euc)

    mds_cos = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
    coords_cos = mds_cos.fit_transform(dist_cos)

    return X_pca, dist_euc, dist_cos, clusters, coords_euc, coords_cos

X_pca, dist_euclidean, dist_cosine, clusters, coords_euclidean, coords_cosine = compute_embeddings(mids_selected)
mids_selected["Cluster"] = clusters
mids_selected["MDS1_euclidean"], mids_selected["MDS2_euclidean"] = coords_euclidean[:,0], coords_euclidean[:,1]
mids_selected["MDS1_cosine"], mids_selected["MDS2_cosine"] = coords_cosine[:,0], coords_cosine[:,1]

top_players = mids_selected.nlargest(25, '90s')['Player'].values
mids_selected['text_label'] = [p if p in top_players else '' for p in mids_selected['Player']]
```

We first visualise the MDS configuration based on Euclidean distances. The MDS algorithm projects players into two dimensions while preserving the pairwise Euclidean distances from the PCA space. Points are coloured by their K-means cluster, and the top 25 players by minutes are labelled for clarity.

```python
fig_euclid = px.scatter(
    mids_selected, x='MDS1_euclidean', y='MDS2_euclidean',
    color=mids_selected['Cluster'].astype(str), text='text_label',
    hover_data={'Player': True,'Squad': True,'Nation': True,'90s': True,'Comp': True,'Age': True,'MDS1_euclidean': False,'MDS2_euclidean': False,'text_label': False},
    labels={'MDS1_euclidean':'Dimension 1','MDS2_euclidean':'Dimension 2','color':'Cluster'},
    title='MDS plot of Midfielders: Euclidean Distance'
)
fig_euclid.update_traces(marker=dict(size=8,line=dict(width=1,color='DarkSlateGrey')),textposition='top center',textfont=dict(size=8.5))
fig_euclid.update_layout(height=800,width=1300)
fig_euclid.show()
```

Next, we visualise the MDS configuration using cosine distances. Cosine distance captures similarity in the direction of the feature vectors rather than magnitude, providing a complementary perspective to Euclidean distance. As before, points are coloured by K-means cluster and the top 25 players by minutes are labelled

```python
fig_cosine = px.scatter(
    mids_selected, x='MDS1_cosine', y='MDS2_cosine',
    color=mids_selected['Cluster'].astype(str), text='text_label',
    hover_data={'Player': True,'Squad': True,'Nation': True,'90s': True,'Comp': True,'Age': True,'MDS1_cosine': False,'MDS2_cosine': False,'text_label': False},
    labels={'MDS1_cosine':'Dimension 1','MDS2_cosine':'Dimension 2','color':'Cluster'},
    title='MDS plot of Midfielders: Cosine Distance'
)
fig_cosine.update_traces(marker=dict(size=8,line=dict(width=1,color='DarkSlateGrey')),textposition='top center',textfont=dict(size=8.5))
fig_cosine.update_layout(height=800,width=1300)
fig_cosine.show()
```

Both MDS plots reveal meaningful structure in the midfielder data. The Euclidean plot reflects differences in raw feature values, so players positioned close together not only perform similar roles but also produce comparable outputs. The cosine plot, by contrast, focuses on the orientation of the feature vectors, highlighting players who carry out similar roles even if their statistical output is lower, essentially showing less efficient versions of similar players.

In both plots, the K-means clusters clearly separate players into groups, which correspond to distinct midfielder archetypes such as creative playmakers, progressive carriers, or defensive-minded midfielders. This demonstrates that the combination of PCA projections, multiple distance metrics, and MDS provides a robust framework for visualising player similarity, understanding playing styles, and identifying nuanced relationships between players that may not be obvious from raw statistics alone
